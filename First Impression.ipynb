```python
import re
from collections import Counter
from nltk.tokenize import RegexpTokenizer

counts=[]
for key,value in trans_training.items():
    tokenizer = RegexpTokenizer(r'\w+')
    text1 = value
    tokens = tokenizer.tokenize(text1)
    text2 =tokens
    nonPunct = re.compile('.*[A-Za-z0-9].*')  # must contain a letter or digit
    filtered = [w for w in text2 if nonPunct.match(w)]
    counts.append(Counter(filtered))
```

```python
counts[0]
```

```python
import numpy as np
unique=[]
for i in counts:
    key=np.array([])
    for k,v in i.items():
        key=np.append(key,k)
    print(key)
    unique.append(np.unique(key))
```
