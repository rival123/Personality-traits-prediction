```python
import re
from collections import Counter
from nltk.tokenize import RegexpTokenizer

counts=[]
for key,value in trans_training.items():
    tokenizer = RegexpTokenizer(r'\w+')
    text1 = value
    tokens = tokenizer.tokenize(text1)
    text2 =tokens
    nonPunct = re.compile('.*[A-Za-z0-9].*')  # must contain a letter or digit
    filtered = [w for w in text2 if nonPunct.match(w)]
    counts.append(Counter(filtered))
```

```python
counts
```

```python
import re
from collections import Counter
from nltk.tokenize import RegexpTokenizer

tot_words=[]
for key,value in trans_training.items():
    tokenizer = RegexpTokenizer(r'\w+')
    text1 = value
    tokens = tokenizer.tokenize(text1)
    tot_words.append(len(tokens))
```

```python
tot_words
```

```python
import numpy as np
unique=[]
for i in counts:
    key=np.array([])
    for k,v in i.items():
        key=np.append(key,k)
    unique.append(key.shape[0])
```

```python
unique
```

```python
dataset = pd.DataFrame({'TotalWordCounts':tot_words,'UniqueWords':unique})
```

```python
dataset
```

```python
df['TotalWordCounts']=dataset['TotalWordCounts']
df['UniqueWords']=dataset['UniqueWords']
```

```python
def find_csv_filenames( path_to_dir, suffix=".csv" ):
    filenames = listdir(path_to_dir)
    return [ filename for filename in filenames if filename.endswith( suffix )]
```
